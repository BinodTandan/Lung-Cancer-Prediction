```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Import and visualize data
df = pd.read_csv("dataset.csv").reset_index(drop=True)
X = df.iloc[:, 2].values  # Select smoking as independent variable
Y = df.iloc[:, 3].values  # Select yellow fingers as dependent variable
df.head()
```

```python
#LINEAR REGRESSION MODEL
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size= 0.3, random_state=0)

sc_X = StandardScaler(with_mean= True)
X_train = sc_X.fit_transform(X_train).reshape(-1, 1)
X_test = sc_X.transform(X_test).reshape(-1, 1)

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)

#Define variables for regression model equation
coeffecients = regressor.coef_
intercept = regressor.intercept_

#Print the equation of the regression model
print(f"The equation of the regression model is: y= {intercept:.2f} + {coeffecients[0]:.2f} * x")

y_pred = regressor.predict(X_test)

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Smoking VS Yellow Fingers (Training set)')
plt.xlabel('Smoking')
plt.ylabel('Yellow Fingers')
plt.show()


```
